### 3. Отобразим список наименований средств, соответствующих двум любым документам с требованиями по сертификации.

```python
import re

def count_commas(str):
    """Функция нахождения количества запятых в строке"""
    
    p = re.compile(',')
    
    # Количество запятых в строке  
    commas_number = len(p.findall(str))
    
    return commas_number
```

```python
commas_number = data['Наименования документов, требованиям которых соответствует средство'].apply(count_commas)
data['docs_number'] = commas_number + 1
```

```python
data['Наименование\r\nсредства (шифр)'][data.docs_number == 2]
```

### 6. Выведем наименования всех средств, которые проходили испытания в частных организациях (форма организации ООО, ЗАО, ПАО, АО).

```python
def check_private_organization(str):
    
    # Возращает 0, если в строке есть хотя бы один из представленных символов
    is_a_privat_org = str.find("ООО ") * str.find("ЗАО ") * str.find("ПАО ") * str.find("АО ")
    
    return is_a_privat_org
```

```python
data['Наименование\r\nсредства (шифр)']/
[data['Испытательная\r\nлаборатория'].apply(check_private_organization) == 0]
```
